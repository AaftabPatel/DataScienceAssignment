1. How to Construct a Bar Graph with Overlay Using Python


```python
import matplotlib.pyplot as plt

a = range(1, 25)
b = a
c = [x+2 for x in b]
d = [x*2 for x in c]

plt.bar(a, d)
plt.bar(a, c)
plt.bar(a, b)
plt.show()
```

2. How to Construct Contingency Tables Using Python


```python
import pandas as pd
import numpy as np

df = pd.read_csv('data.csv', nrows=5)

data_crosstab = pd.crosstab(df['Duration'],df['Pulse'],margins = False)
print(data_crosstab)
```

4. How to Perform Binning Based on Predictive Value Using Python

```python
import pandas as pd

#create DataFrame
df = pd.DataFrame({'points': [4, 4, 7, 8, 12, 13, 15, 18, 22, 23, 23, 25],
                   'assists': [2, 5, 4, 7, 7, 8, 5, 4, 5, 11, 13, 8],
                   'rebounds': [7, 7, 4, 6, 3, 8, 9, 9, 12, 11, 8, 9]})

#view DataFrame
print(df)

#perform data binning on points variable
df['points_bin'] = pd.qcut(df['points'], q=5)

#view updated DataFrame
print("\n",df)

#count frequency of each bin
df['points_bin'].value_counts()
```
